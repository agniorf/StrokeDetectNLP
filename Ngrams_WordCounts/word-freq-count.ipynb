```python
import os
import string
import nltk
from nltk.tokenize import RegexpTokenizer
import pandas as pd

file = "~/Dropbox (Partners Healthcare)/MVP/RPDR/Cerebral Edema 2014-2018/Data/Processed/Unstructured/Rad_processed.txt"
```

```python
with open(os.path.expanduser(file), 'r') as f:
    text = f.read()
```

```python
text = text.lower()
# text = text.translate(str.maketrans('','',string.punctuation))
```

```python
text[:1000]
```

```python
tokenizer = RegexpTokenizer(r'\w+')
counts = {}
for token in tokenizer.tokenize(text):
    if token in counts:
        counts[token] += 1
    else:
        counts[token] = 1
```

```python
counts
```

```python
to_df = {'counts':list(counts.values()), 'words':list(counts.keys())}
```

```python
df = pd.DataFrame(data=to_df)
```

```python
df.sort_values(by='counts', ascending=False)
```

```python
df.to_csv('radiology_word_counts.csv')
```
