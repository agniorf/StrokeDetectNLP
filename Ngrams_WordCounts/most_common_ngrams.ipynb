```python
import os
import string
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.util import ngrams
from collections import Counter

import pandas as pd

file = "~/Dropbox (Partners Healthcare)/MVP/RPDR/Cerebral Edema 2014-2018/Data/Processed/Unstructured/Rad_processed.txt"
```

```python
with open(os.path.expanduser(file), 'r') as f:
    text = f.read()
```

```python
text = text.lower()
tokens = nltk.word_tokenize(text)
```

```python
bigrams = ngrams(tokens,2)
bg = list(bigrams)
Counter([i for i in bg if all(j not in string.punctuation for j in i)]).most_common()
```

```python
trigrams = ngrams(tokens,3)
bg = list(trigrams)
Counter([i for i in bg if all(j not in string.punctuation for j in i)]).most_common()
```

```python
fourgrams = ngrams(tokens,4)
bg = list(fourgrams)
Counter([i for i in bg if all(j not in string.punctuation for j in i)]).most_common()
```
